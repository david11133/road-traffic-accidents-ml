### Accident Severity Distribution
This count plot shows how the accidents are distributed by severity. We notice that the dataset is highly imbalanced, with a majority of accidents being "Slight Injury." 
This might affect model performance, and we may need to address it later using techniques like SMOTE or class weighting.


```python
# Checking the Distribution of Accident Severity
sns.countplot(data=df, x='Accident_severity', palette='Set2')
plt.title('Distribution of Accident Severity')
plt.xticks(rotation=45)
plt.show()
```

### Class Imbalance Check
Here, we calculate the proportion of each class. The data is highly imbalanced with the "Slight Injury" class making up 85% of the cases, while "Fatal Injury" is much rarer. This could be important for model evaluation and performance.

```python
df['Accident_severity'].value_counts(normalize=True)
```

### Analyzing Accident Severity vs. Number of Vehicles Involved
This boxplot helps us understand the relationship between accident severity and the number of vehicles involved. Visualizing such relationships can provide insights into how features influence the target variable.

```python
# Analyzing the Relationship Between 'Accident_severity' and Other Features
sns.boxplot(x='Accident_severity', y='Number_of_vehicles_involved', data=df)
plt.title('Accident Severity vs. Number of Vehicles Involved')
plt.show()
```

```python
# Comparing Sex_of_driver with Accident Severity
sns.countplot(data=df, x='Sex_of_driver', hue='Accident_severity')
plt.title('Sex of Driver vs. Accident Severity')
plt.show()
```
